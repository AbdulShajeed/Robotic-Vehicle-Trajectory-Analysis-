```python
#understanding the correlation between columns:
shuffled_df['label'] = 0
shuffled_df['label'].mask(shuffled_df['Status']=='LEFT', 1, inplace=True)
shuffled_df['label'].mask(shuffled_df['Status']=='STRAIGHT', 2, inplace=True)
shuffled_df.corr()
```

```python
sns.pairplot(df,hue="Status")
```

```python
Y = shuffled_df['label']
specs = df['Status']
X = shuffled_df.drop(["label", "Status"], axis="columns")
from sklearn.model_selection import train_test_split 
accuracy_data=[]
#, random_state=42
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.10, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics 

clf = LogisticRegression()
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
accuracy_data.append({"name":"Logistic Regression","accuracy":metrics.accuracy_score(y_test, y_pred)})
print(metrics.accuracy_score(y_test, y_pred))
print(metrics.classification_report(y_test, y_pred))
```

```python
from sklearn.svm import SVC
model = SVC(C=72)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy_data.append({"name":"SVM","accuracy":metrics.accuracy_score(y_test, y_pred)})
print(metrics.accuracy_score(y_test, y_pred))
print(metrics.classification_report(y_test, y_pred))
```

```python
print(model.score(x_test, y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier 
```

```python
model = RandomForestClassifier(n_estimators=2)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
accuracy_data.append({"name":"Random Forest","accuracy":metrics.accuracy_score(y_test, y_pred)})
print(classification_report(y_test,y_pred))
```

```python
rf = RandomForestClassifier(random_state =42, n_jobs=-1 )
params = {'max_depth': range(1,10),
          'min_samples_leaf': range(1,10),
          'n_estimators': range(1,25)}
```

```python
from sklearn.model_selection import GridSearchCV
grid_search  =GridSearchCV(estimator = rf,
                           param_grid=params,
                           cv = 4, verbose =3)




```

```python
grid_search.fit(x_train, y_train)
```

```python
rf = RandomForestClassifier(random_state =42, n_jobs=-1 )
params = {'max_depth': range(1,10),
          'min_samples_leaf': range(1,10),
          'n_estimators': range(1,25)}
```

```python
# Print the best hyperparameters found
print(grid_search.best_params_)

# Print the classification report for the best estimator
best_estimator = grid_search.best_estimator_
y_pred = best_estimator.predict(x_test)
report = classification_report(y_test, y_pred)
print(report)
accuracy_data.append({"name":"Grid Search CV","accuracy":metrics.accuracy_score(y_test, y_pred)})
```

```python
from sklearn.ensemble import BaggingClassifier
bagging = BaggingClassifier(n_estimators=100, random_state=42)
bagging.fit(x_train, y_train)
y_pred = bagging.predict(x_test)
accuracy_data.append({"name":"Bagging Classifier","accuracy":metrics.accuracy_score(y_test, y_pred)})
print(classification_report(y_test, y_pred))
```

```python
print(accuracy_data)
xaxis = [sub['name'] for sub in accuracy_data ]
yaxis = [ sub['accuracy'] for sub in accuracy_data ]
plt.plot(xaxis,yaxis)  # Plot the chart
plt.ylim(0, 1)
plt.show()  

fig, ax = plt.subplots()
width = 0.4

ax.bar(xaxis, yaxis, width, color = "green")
rects = ax.patches
for i, v in zip(rects,yaxis):
    height = i.get_height()
    ax.text( i.get_x() + i.get_width() / 2, height, str(round(v,2)), ha="center", va="bottom")
plt.show()
```
